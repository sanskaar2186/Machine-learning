```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/spam.csv')
df
```

```python
df['Spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df.head()
```

```python
X=df['Message']
y=df['Spam']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer


vector=TfidfVectorizer()

```

```python
X_train_count=vector.fit_transform(X_train)
X_test_count=vector.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
y_pred=model.predict(X_test_count)
y_pred
```

```python
y_test
```

```python
accuracy = model.score(X_test_count, y_test)
print("Accuracy:", accuracy)
```

```python
def predict_spam(message, vector, model):
    message_vector = vector.transform([message])
    prediction = model.predict(message_vector)

    if prediction == 1:
        print("🚨 This is SPAM.")
    else:
        print("✅ This is NOT spam.")
```

```python
predict_spam("Congratulations! You won a free iPhone!", vector, model)
predict_spam("Hey, are we still meeting at 3?", vector, model)
```
